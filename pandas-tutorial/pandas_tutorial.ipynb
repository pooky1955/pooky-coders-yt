# Column Manipulation
A Dataframe is made of rows and columns.
- `df['foobar']`: accesses the column `foobar` in df  
- `df.iloc[0]`: gets the first row
- `df.iloc[i]`: gets the ith row
- `df.iloc[[i,j,k]]`: gets the ith,jth, and kth rows
- `df['foobar'] = some_values`: updates/creates the column `foobar`
- `pd.to_datetime(df['foobar'])`: converts a column's type to datetime
- `pd.to_numeric(df['foobar'])`: converts a column's type to numerical
- `pd.to_categorical(df['foobar'])`: converts a column's type to categorical
